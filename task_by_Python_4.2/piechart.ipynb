```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
R_names = pd.read_csv('names/yob{}.txt'.format(1950),names=['Name','Gender','Count'])
```

```python
R_names_pie = R_names.query('Name.str[0] == "R"').groupby('Name').sum().sort_values(by='Count', ascending=False).head(20)
```

```python
R_names_pie.plot.pie(y='Count')
```
